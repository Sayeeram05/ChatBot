```python
%run Model.ipynb import NeuralNet
%run NLTK.ipynb import Tokenize,BagOfWord,Lower


FILE = "data.pth"
data = torch.load(FILE)

input_size = data["input_size"]
hidden_size = data["hidden_size"]
output_size = data["output_size"]
all_words = data['all_words']
tags = data['tags']
model_state = data["model_state"]
```

```python
import json
with open("TestData.json", "r") as D:
    intents = json.load(D)
intents
```

```python
Test = []
for i in intents["Data"]:
    #print(i["tag"])
    for j in i["patterns"]:
        #print(j)
        Test.append([j,i["tag"]])
```

```python
len(Test)
```

```python
import random
TestSize = 200
Test100 = []

for i in range(TestSize):
    Test100.append(Test[random.randrange(0,len(Test))])
```

```python
Test100
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
model = NeuralNet(input_size, hidden_size, output_size).to(device)
model.load_state_dict(model_state)
model.eval()
```

```python
Correct = 0
Wrong = 0
for pattern,Y in Test100:
    Words = Tokenize(pattern)
    X = BagOfWord(Words, all_words)
    X = X.reshape(1, X.shape[0])
    X = torch.from_numpy(X).to(device)


    output = model(X)
    _, predicted = torch.max(output, dim=1)

    tag = tags[predicted.item()]

    
    if(Y == tag):
        Correct += 1

    else:
        Wrong += 1
        print(f"Pattern : {pattern}")
        print(f"Bot = {tag},Dataset = {Y}")

print()
print("Correct : ",Correct)
print("Wrong : ",Wrong)
print("Accyracy : ",Correct/TestSize)
```
